```python
#highest average votting
data.groupby('Released_Year')['No_of_Votes'].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(12,6),dpi=100)
data.groupby('Released_Year')['No_of_Votes'].mean().sort_values(ascending=False).head(15).plot(kind='bar',color='green')
plt.xlabel('Year',size=20)
plt.ylabel('Votes',size=20)
plt.title('Top 15 average votes by year',size=20)
plt.show()
```

```python
data['Gross'] = data['Gross'].fillna(0)

data.pivot_table(values=['Gross','No_of_Votes','Meta_score'], index='Released_Year', aggfunc='mean').sort_values(by=['Meta_score','No_of_Votes'],ascending=False).plot()
```

```python
# average rating for each director
data.groupby('Director')['IMDB_Rating'].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(12,6),dpi=100)
data.groupby('Director')['IMDB_Rating'].mean().sort_values(ascending=False).head(10).plot(kind='bar')
plt.xlabel('Director',size=15)
plt.ylabel('Rating',size=15)
plt.title('Top 10 average rating for each directo',size=15)
plt.show()
```

```python
data['Runtime']=data['Runtime'].str.replace('min','')
```

```python
data['Runtime']=pd.to_numeric(data['Runtime'])
```

```python
# top lenghty movies
top_10_movies = data.sort_values(by='Runtime',ascending=False)[['Series_Title','Runtime']].head(10)
```

```python
sns.barplot(x='Series_Title',y='Runtime',data =top_10_movies)
plt.title('Top 10 length movies')
plt.xticks(rotation=90)
```

```python
# cout of movies/Tv shows in each genre
genreCount= data.Genre.value_counts()

genreCount
```

```python
# Bar Chart visualization
import matplotlib.pyplot as mpl

movies=genreCount[0:10].index
count=list(genreCount.iloc[:10])

mpl.barh(movies, count)
mpl.ylabel("movies/shoes genre")
mpl.xlabel("Count")
mpl.title("Top 10 movie/shows genre vs. their counts")

mpl.show()
```
